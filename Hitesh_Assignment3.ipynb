```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
iris = sns.load_dataset("iris")
```

```python
plt.scatter(iris["sepal_width"],iris["sepal_length"])
plt.xlabel("sepal_width")
plt.ylabel("sepal_length")
plt.title("Scatter plot")
plt.show()
```

```python
sns.scatterplot(x="sepal_width",y="sepal_length",data=iris)
```

```python
plt.bar(iris["petal_width"],iris["petal_length"])
plt.xlabel("petal_length")
plt.ylabel("petal_width")
plt.title("Bar Graph")
plt.show()
```

```python
sns.lineplot(x="sepal_width",y="sepal_length", data = iris)
plt.show()
```

```python
sns.lineplot(x="petal_width",y="petal_length", data = iris)
plt.show()
```

```python
plt.subplot(2,2,1)
plt.plot(iris["sepal_width"],iris["sepal_length"] , color = "blue")
plt.subplot(2,2,2)
plt.bar(iris["sepal_width"],iris["sepal_length"] , color = "red")
plt.subplot(2,2,3)
plt.scatter(iris["sepal_width"],iris["sepal_length"] , color = "purple")
plt.show()
```
